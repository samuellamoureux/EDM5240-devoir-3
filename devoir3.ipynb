# Données sur les demandes d'accès à l'info

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
md = pd.read_csv("ati.csv")
```

Cette première commande de base indique seulement le nombre d'années représentées par le fichier. On voit qu'il recouvre seulement les demandes de 2015 à 2017. 

```python
md.year.value_counts()
```

Les deux commandes ci-dessous, dont le premier tableau, nous informent des statistiques sur les réponses données aux demandes d'accès à l'information. La majorité des réponses sont des communications partielles(8653). Il y a ensuite beaucoup moins de communications totales(2907), et un peu de demandes avec aucun document existant(883). 


```python
md.disposition.value_counts()
```

```python
md.disposition.value_counts().plot(kind="pie")
```
